## Question 1: Which of the country recorded the highest sales and why?

```python
df.groupby(['Country']).sum()
```

```python
# NOW LETS VISUALIZE THIS TO GET MORE INSIGH FROM THIS SOLUTION
```

```python
country_group =[ pair for pair, df in df.groupby(['Country'])]
Sales = df.groupby(['Country']).sum()['sales']
plt.bar(country_group,Sales ,color = 'r')
plt.xticks(country_group, rotation ='vertical')
plt.xlabel('country')
plt.ylabel('sales in ($)')
plt.show()
```

#### Answer: The United Kingdom recorded the highest sales reasons being that the central office could be in the UK

### Question 2: Which Day recorded the highest sale and why?

```python
df.groupby(['Day']).count()['sales']
```

```python
# ploting the visuals will help us gain better insight and and understanding
```

```python
days_group =[ key for key, df in df.groupby(['Day'])]
SALES= df.groupby(['Day']).count()['sales']
plt.bar(days_group, SALES)
plt.xticks(days_group, rotation ='vertical')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

ANS: The highest sale was recorded on the 6th of december and the lowest was recored on the 31st of december reason being 
that at the end of the year most people are probably done buying all the needed since christmas celebration would be over by then.
